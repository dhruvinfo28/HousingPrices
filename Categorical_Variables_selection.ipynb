```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
y =data.SalePrice
```

```python
data = data.select_dtypes(include='object')
data.info()
```

```python
#Columns with missing values
col_missing_values = []
for i in data.columns:
    if data[i].isnull().any():
        col_missing_values.append(i)
col_missing_values
```

```python
data.drop(col_missing_values,axis=1,inplace=True)
data.info()
```

```python
#Now ill use onehot encoding 
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore',sparse=False)
```

```python
X_categorical = pd.DataFrame(encoder.fit_transform(data))
X_categorical.index = data.index
```

```python
X_categorical
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid = train_test_split(X_categorical,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=1000)
rfr.fit(X_train,y_train)
predict = rfr.predict(X_valid)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_valid,predict)
mae
```
